## Grouping

```python
# Let's calculate the mean number of vaccinations per country

df_mean_vaccinations_per_country = df.groupby('country').agg(mean_daily_vaccinations = ('daily_vaccinations', 'mean'))

df_mean_vaccinations_per_country.sort_values(by='mean_daily_vaccinations', ascending=False).reset_index()
```

```python
# Let's calculate the global number of vaccinations on any given day

df_total_vaccinations_per_day = df.groupby('date').agg(total_vaccinations = ('daily_vaccinations' , 'sum')).reset_index()

df_total_vaccinations_per_day
```

```python
from matplotlib import pyplot as plt
```

```python
plt.plot(df_total_vaccinations_per_day.date, df_total_vaccinations_per_day.total_vaccinations)

plt.xlabel('Days')
plt.ylabel('Number of Global vaccinations')

plt.show()
```

## Merging

```python
df_countries = pd.read_csv('countries.csv')
```

```python
df_countries.sample(15)
```

```python
df_countries = df_countries.loc[:, ['alpha-3', 'region']]
df_countries
```

### Renaming columns

```python
df_countries = df_countries.rename(columns={'alpha-3' : 'iso_code',
                                            'region' : 'continent'})
df_countries
```

```python
df_covid_continents = df_countries.merge(df,
                                         on='iso_code',
                                         how='inner')

df_covid_continents
```

```python
# Question 1

df_1 = df_covid_continents.loc[:, ['continent', 'country', 'date', 'people_fully_vaccinated', 'daily_vaccinations', 'daily_vaccinations_per_million']]

df_1
```

```python
df_quest_2 = df_1.loc[df_1.daily_vaccinations_per_million != 0, :]

df_quest_2.loc[:, 'population_in_millions'] = df_quest_2.daily_vaccinations / df_quest_2.daily_vaccinations_per_million


df_country_population = df_quest_2.groupby('country').agg(continent=('continent', 'first'),
                                                          population=('population_in_millions', 'mean')).reset_index()

df_continent_population = df_country_population.groupby('continent').agg(population=('population', 'sum'))

df_continent_population
```

```python
# Question 3

df_quest_3 = df_1.groupby('continent').agg(n_countries = ('country', 'nunique'))

df_quest_3
```

```python
# Question 4

df_quest_4 = df_1.loc[df_1.continent == 'Africa', :]

# Either calculate the number of fully vaccinated people for eah country and then find the country 

# Or just look for the number corresponding to the max since the value is incremental


df_quest_4.loc[df_quest_4.people_fully_vaccinated == df_quest_4.people_fully_vaccinated.max()]
```

```python
# Question 5

df_quest_5 = df_1.loc[df_1.date.dt.year == 2022]

df_quest_5 = df_quest_5.groupby('continent').daily_vaccinations.sum()

df_quest_5
```

```python
# Question 6

df_quest_6 = df_1.loc[(df_1.country=='Morocco') & (df_1.date.dt.month == 12) & (df_1.date.dt.year == 2021)]

# or e can use between

median_daily_vaccinations = df_quest_6.daily_vaccinations.median()

print(f"The median number of daily vaccinations in Morocco is : {median_daily_vaccinations}")
```

```python
import numpy as np

global_population = df_continent_population.population.sum() * 1e6

global_population = np.round(global_population, 0)

print(f"The global population is : {global_population}")

# Lets calculate the number of fully vaccinated people
# Let's take the max value for each country and then sum

df_fully_vaccinated = df_1.groupby('country').people_fully_vaccinated.max()

fully_vaccinated_people = df_fully_vaccinated.sum()

percentage_fully_vaccinated = fully_vaccinated_people / global_population * 100
print(f"The percentage of fully vaccinated people is : {np.round(percentage_fully_vaccinated, 2)}%")
```
